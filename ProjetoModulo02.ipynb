Importação de bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import seaborn as sns
```

## Importação das bases de dados

```python
# Importação da base de dados de Doses Aplicadas
doses_aplicadas = pd.read_csv("https://github.com/cmattosr/bootcampDS_Projeto02/raw/main/dados/Doses_aplicadas%20por%20Anom%C3%AAs%20segundo%20Unidade%20da%20Federa%C3%A7%C3%A3o%20Febre%20Amarela.csv", sep=";", encoding="ISO-8859-1", decimal=",", thousands=".", engine="python")
doses_aplicadas
```

```python
# Importação da base de dados de Internações Hospitalares
internacoes = pd.read_csv("https://github.com/cmattosr/bootcampDS_Projeto02/raw/main/dados/Interna%C3%A7%C3%B5es%20por%20Anom%C3%AAs%20processamento%20segundo%20Unidade%20da%20Federa%C3%A7%C3%A3o%20Febre%20Amarela.csv", skiprows=4, skipfooter=6, sep=";", encoding="ISO-8859-1",decimal=",", thousands=".", engine="python")
internacoes
```

```python
# Importação da base de dados de Óbitos
obitos = pd.read_csv("https://github.com/cmattosr/bootcampDS_Projeto02/raw/main/dados/%C3%93bitos%20por%20Anom%C3%AAs%20processamento%20segundo%20Unidade%20da%20Federa%C3%A7%C3%A3o%20Febre%20Amarela.csv", skiprows=4, skipfooter=6, sep=";", encoding="ISO-8859-1",decimal=",", thousands=".", engine="python")
obitos
```

```python
# Importação da base de dados de Cobertura Vacinal
cobertura_vacinal = pd.read_csv("https://github.com/cmattosr/bootcampDS_Projeto02/raw/main/dados/Coberturas%20Vacinais%20por%20Ano%20segundo%20Unidade%20da%20Federa%C3%A7%C3%A3o%20Febre%20Amarela.csv", sep=";", encoding="ISO-8859-1",decimal=",", thousands=".", engine="python")
cobertura_vacinal
```

## Limpeza dos dados importados

### Doses Aplicadas

```python
# Todas as colunas do DataFrame, e a quantidade de linhas e colunas
print(*doses_aplicadas.columns)
print(doses_aplicadas.shape)
```

```python
# Definição dos estados como índice 
doses_aplicadas = doses_aplicadas.set_index("Unidade da Federação")
# Seleção dos anos entre 2009 e 2020
doses_aplicadas = doses_aplicadas.loc[:, "2008/Jan" : "2020/Dez"]
doses_aplicadas
```

```python
# Verificar se há valor nulo
doses_aplicadas = doses_aplicadas.replace('-', np.nan)
doses_aplicadas.T.info()
```

### Internações Hospitalares (essa base não tem todos os meses do anos, e nem todos os estados tiveram internações por Febre Amarela)

```python
# Todas as colunas do DataFrame, e a quantidade de linhas e colunas
print(*internacoes.columns)
print(internacoes.shape)
```

```python
# Definição dos estados como índice 
internacoes = internacoes.set_index("Unidade da Federação")
# Seleção dos anos entre 2009 e 2020
internacoes = internacoes.loc[:, "2008/Jan" : "2020/Out"]
internacoes
```

```python
internacoes.T.info()
```

```python
#transformar valores em números
internacoes = internacoes.replace('-', np.nan)
internacoes = internacoes.astype("float64")
internacoes.T.info()
```

### Obitos (essa base não tem todos os meses do anos, e nem todos os estados tiveram óbitos por Febre Amarela)

```python
# Todas as colunas do DataFrame, e a quantidade de linhas e colunas
print(*obitos.columns)
print(obitos.shape)
```

```python
# Definição dos estados como índice 
obitos = obitos.set_index("Unidade da Federação")
# Seleção dos anos entre 2009 e 2020 (na base não tem os meses de Nov e Dez de 2020)
obitos = obitos.loc[:, "2008/Mar" : "2020/Mai"]
obitos
```

```python
obitos.T.info()
```

```python
#transformar valores em números
obitos = obitos.replace('-', np.nan)
obitos = obitos.astype("float64")
obitos.T.info()
```

### Cobertura Vacinal

```python
# Definição dos estados como índice 
cobertura_vacinal = cobertura_vacinal.set_index("Unidade da Federação")
# Seleção dos anos entre 2009 e 2020
cobertura_vacinal = cobertura_vacinal.loc[:, "2008" : "2020"]
cobertura_vacinal
```

## Análise Exploratória

```python
# Vamos verificar o total de doses aplicadas no país entre 2008 e 2020
plt.figure(figsize=(15,10))
ax = doses_aplicadas.loc["Total", :].plot()
plt.xticks(rotation=45)
plt.title("Doses Aplicadas em todo o território nacional entre 2008 e 2020", size=14)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
ax.xaxis.set_major_locator(ticker.IndexLocator(base=12, offset=0))
plt.show()
```

```python
# Verificamos uma diferença significativa entre os totais de doses ano a ano

# vamos imprimir todos os valores totais deste Dataframe para fazermos uma verificação extra
print(doses_aplicadas.loc["Total", :].to_string())
```

```python
# Vamos verificar o total de internações no país entre 2008 e 2020
plt.figure(figsize=(15,10))
ax = internacoes.loc["Total", :].plot()
plt.xticks(rotation=45)
plt.title("Internações em todo o território nacional entre 2008 e 2020", size=14)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.show()
```

```python
# É possível identificar maior atividade entre os anos de 2016 e 2019, vamos plotar esse período
plt.figure(figsize=(15,10))
ax = internacoes.loc["Total","2016/Nov": "2018/Out"].plot()
plt.xticks(rotation=45)
plt.title("Internações em todo o território nacional entre 2016/Nov e 2018/Out", size=14)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
ax.xaxis.set_major_locator(ticker.IndexLocator(base=1, offset=0))
plt.show()
```

É possível identificar picos de internações devido a febre amarela entre 2016/Nov e 2018/Out. Vamos voltar ao gráfico de doses aplicadas plotando o período entre 2016/Nov e 2018/Outespecial a esse período.

```python
# Vamos voltar ao gráfico de doses aplicadas plotando o período entre 2016/Nov e 2018/Out
plt.figure(figsize=(15,10))
ax = doses_aplicadas.loc["Total","2016/Nov": "2018/Out"].plot()
plt.xticks(rotation=45)
plt.title("Doses Aplicadas em todo o território nacional entre 2016/Nov e 2018/Out", size=14)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
ax.xaxis.set_major_locator(ticker.IndexLocator(base=1, offset=0))
plt.show()
```

```python
# Vamos continuar a análise exploratória
# Vamos verificar o total de óbitos no país entre 2008 e 2020
plt.figure(figsize=(15,10))
ax = obitos.loc["Total", :].plot()
plt.xticks(rotation=45)
plt.title("Doses Aplicadas em todo o território nacional entre 2008 e 2020", size=14)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
ax.xaxis.set_major_locator(ticker.IndexLocator(base=1, offset=0))
plt.show()
```

```python
# É possível identificar os picos no mesmo período que identificamos nas doses aplicadas e nas internações
# vamos plotar o período entre 2012/Mai e 2018/Out porque não há informção entre 2012/Jun e 2016/Dez

plt.figure(figsize=(15,10))
ax = obitos.loc["Total", "2012/Mai" : "2018/Nov"].plot()
plt.xticks(rotation=45)
plt.title("Doses Aplicadas em todo o território nacional entre 2008 e 2020", size=14)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
ax.xaxis.set_major_locator(ticker.IndexLocator(base=1, offset=0))
plt.show()
```

A cobertura vacinal é dada em percentual e apenas os valores anuais, sem distinção entre os meses do ano

```python
# Vamos verificar se há picos no Dataframe cobertura_vacinal
plt.figure(figsize=(15,10))
ax = cobertura_vacinal.loc["Total", :].plot()
plt.xticks(rotation=45)
plt.title("Cobertura Vacinal em todo o território nacional entre 2008 e 2020", size=14)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
ax.xaxis.set_major_locator(ticker.IndexLocator(base=1, offset=0))
plt.show()
```

```python
# Identificamos o pico a partir do ano de 2016
plt.figure(figsize=(15,10))
ax = cobertura_vacinal.loc["Total", "2016":].plot()
plt.xticks(rotation=45)
plt.title("Cobertura Vacinal em todo o território nacional entre 2016 e 2020", size=14)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
ax.xaxis.set_major_locator(ticker.IndexLocator(base=1, offset=0))
plt.show()
```

### Através da análise exploratória foi possível identificar que o período entre 2016/Nov e 2018/Out tem um comportamento muito diferente, tendo dois picos muito mais altos do que os valores anteriores e posteriores a ele. A partir de agora vamos nos concentrar neste período, e tentaremos identificar se há algum estado com comportamento diferente dos outros
